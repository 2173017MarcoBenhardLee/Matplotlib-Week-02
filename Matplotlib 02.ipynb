Pengaturan Format pada plot
Nilai default untuk parameter ketiga = 'b-'artinya blue garis
Pemanggilan fungsi axis() menyertakan sebuah list[xmin, xmax, ymin, ymax]

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') #parameter ketiga pengatur bentuk marker yang dimaksud adalah karakter o dan bukan nol
                                    #r- = garis, r* = bintang
plt.axis([0,6,0,20])
plt.show()
```

Multi plot dalam satu pemanggilan fungsi plot()
t = np.arrange(nilai min, nilai max, interval)

```python
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t,t,'r--') #menggunakan t sebagai nilai x dan y 
plt.plot(t,t**2,'bs') #merepresentasikan data kuadrat, 'bs'= marker kotak biru
plt.plot(t,t**3,'g^') #merepresentasikan data kubik, 'g^'= marker segitiga hijau
plt.show()
```

Melakukan multi plot dalam satu pemanggilan fungsi plot()

```python
plt.plot(t, t, 'r--',t,t**2,'bs-',t,t**3,'g^')
plt.show()
```

PLotting dengan keywords
Melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument(kwargs) seperti pada 'dictionary' dan 'Pandas DataFrame'

```python
data = {'a' : np.arange(50), 'c' : np.random.randint(0,50,50), 'd' :np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a','b', c = 'c', s = 'd', data=data) #c ='c' representasi melalui warna, s ='d' representasi ukuran, semakin besar nilai d semakin besar ukuran representasi poinnya
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
